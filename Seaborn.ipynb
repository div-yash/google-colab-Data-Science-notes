Seaborn :
 Seaborn is a library that uses Matplotlib underneath to plot graphs. It will be used to visualize random distributions.



*   Data Visualization Library

ts syntax is concise, and it offers very attractive default themes. It's an ideal tool for statistical visualization. It is used to summarize data in visualizations and data distribution. Furthermore, Seaborn is better integrated than Matplotlib for working with Pandas data frames.



```python
#importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note : Seaborn has some built in datasets


```python
#total bill vs tip dataset
tips =  sns.load_dataset('tips')

print(tips)
```

```python
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
# visualise the data
sns.relplot(data=tips,x='total_bill', y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()

house=pd.DataFrame(housing.data,columns=housing.feature_names)
house['PRICE']=housing.target
```

```python
print(housing)
```

```python
house.head()
```

Distribution PLOT:


```python
sns.displot(house['PRICE'])
```

Correlation :

1. postive
2. negative


Heat MAP:

```python
correlation=house.corr()
```

```python
# constructing a heat map
plt.figure(figsize=(10,10))
#           items to plot    color bar       square shaped   no. of floating points
sns.heatmap(correlation,     cbar=True,    square=True,       fmt='.1f',
#   anotation(values)          size of the values    color
    annot=True , annot_kws={'size':8},         cmap='Greens' )
```
