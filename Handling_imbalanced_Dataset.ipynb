Imbalanced Dataset:
  a dataset with an unequal class distribution .

```python
#importing the dependencies
import pandas as pd
import numpy as np
```

```python
#load the dataset to pandas datafrme
credit_card_data=pd.read_csv('/content/credit_data.csv')
```

```python
#print the first five rows of dataframe
print(credit_card_data.head())
```

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#determine the disstribution of the two classes
credit_card_data['Class'].value_counts()
```

This is a highly imbalanced dataset

0 ---> legit transaction

1 ---> fraud transaction

```python
#seperating the legit and fraudalent transactions
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit)
print(legit.shape)
```

```python
print(fraud)
print(fraud.shape)
```

Under-sampling

Build a sample dataset containing similar distribution of legit and fraudulent transactions


Number of fraudulent tranctions -> 134

```python
legit_sample=legit.sample(n=134)
```

```python
print(legit_sample.shape)
```

Concatenate the two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud],axis=0) # concat row wise
```
